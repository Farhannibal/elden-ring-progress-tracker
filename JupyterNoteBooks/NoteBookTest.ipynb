```python
SLOT_START_INDEX = 0x310;
SLOT_LENGTH = 0x280000;
SAVE_HEADERS_SECTION_START_INDEX = 0x19003B0;
SAVE_HEADERS_SECTION_LENGTH = 0x60000;
SAVE_HEADER_START_INDEX = 0x1901D0E;
SAVE_HEADER_LENGTH = 0x24C;
CHAR_ACTIVE_STATUS_START_INDEX = 0x1901D04;

CHAR_NAME_LENGTH = 0x22;
CHAR_LEVEL_LOCATION = 0x22;
CHAR_PLAYED_START_INDEX = 0x26;

savePath = 'C:/Google Drive/dev/EldenRingSaveViewer/EldenRingProgressTracker/elden-ring-progress-tracker/src/files/testSave/ER0000.sl2'
```

```python
class dataObject(object):
    pass

def LoadData(data, slotIndex):
    saveDataObject = dataObject()
    try:
        saveDataObject.index = slotIndex
        data.seek(CHAR_ACTIVE_STATUS_START_INDEX)
        saveDataObject.active =  list(data.read(10))[0 + slotIndex] == 1 if True else False
        data.seek(SAVE_HEADER_START_INDEX + (slotIndex * SAVE_HEADER_LENGTH))
        saveDataObject.characterName = data.read(CHAR_NAME_LENGTH).decode('utf_16')
        data.seek(SAVE_HEADER_START_INDEX + (slotIndex * SAVE_HEADER_LENGTH))
        saveDataObject.CharacterLevel = list(data.read())[CHAR_LEVEL_LOCATION]
        data.seek(SAVE_HEADER_START_INDEX + (slotIndex * SAVE_HEADER_LENGTH) + CHAR_PLAYED_START_INDEX)
        saveDataObject.SecondsPlayed = int.from_bytes(data.read(4), "little")
        data.seek(SLOT_START_INDEX + (slotIndex * 0x10) + (slotIndex * SLOT_LENGTH))
        saveDataObject.saveData = list(data.read(SLOT_LENGTH))
        data.seek(SAVE_HEADER_START_INDEX + (slotIndex * SAVE_HEADER_LENGTH))
        saveDataObject.headerData = list(data.read(SAVE_HEADER_LENGTH))
        return saveDataObject
    except:
        print('Data Incompatible')
        return None;
```

```python
SaveData = LoadData(open(savePath, 'rb'),0)
```

```python
print("Index: " + str(SaveData.index))
print("Active: " + str(SaveData.active))
print("Character Name: " + str(SaveData.characterName))
print("Character Level: " + str(SaveData.CharacterLevel))
print("Time Played: " + str(SaveData.SecondsPlayed/(60*60)))
```
